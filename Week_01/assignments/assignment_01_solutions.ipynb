## Task 4: Data Exploration

- Load and visualize the example EM image from the `raw_data` folder using appropriate libraries for microscopy data.
- Compute basic image statistics such as mean intensity and standard deviation.
- Plot a histogram of pixel intensities.
- Add brief text cells explaining what you observe in the plots.

````python
# Your code here
import matplotlib.pyplot as plt
import hyperspy.api as hs

# Load the EM image
img = hs.load('raw_data/example_EM_Image.tif')

# Display basic information
print(f"Image shape: {img.data.shape}")
print(f"Image data type: {img.data.dtype}")

# Plot the image
img.plot()
````

````python
# Compute basic image statistics
import numpy as np

mean_intensity = np.mean(img.data)
std_intensity = np.std(img.data)

print(f"Mean intensity: {mean_intensity}")
print(f"Standard deviation: {std_intensity}")

# Plot histogram of pixel intensities
plt.figure(figsize=(8, 6))
plt.hist(img.data.flatten(), bins=50, alpha=0.7)
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.show()
````
